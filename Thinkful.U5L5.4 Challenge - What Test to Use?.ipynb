### Did people become less trusting from 2012 to 2014? Compute results for each country in the sample.

```python
cntry_2012 = df_2012.groupby('cntry')
cntry_2014 = df_2014.groupby('cntry')

trst_2012 = cntry_2012['ppltrst'].mean()
trst_2014 = cntry_2014['ppltrst'].mean()

print(trst_2014 - trst_2012)
```

**Generally speaking, people seemed to become more trusting from 2012 to 2014. Sweden showed the greatest increase in trust from 2012 to 2014 (0.199210), followed by Denmark (0.142857), Switzerland (0.073739), and the Czech Republic (0.062138). The two countries whose trust scores decreased were Norway (-0.050685) and Spain (-0.199210).**

### Did people become happier from 2012 to 2014? Compute results for each country in the sample.

```python
happy_2012 = cntry_2012['happy'].mean()
happy_2014 = cntry_2014['happy'].mean()

print(happy_2014 - happy_2012)
```

**Again, it appears that, generally speaking, as a whole people felt happier from 2012 to 2014. Denmark showed the greatest increase (0.428571), followed by the Czech Republic (0.143213), Sweden (0.039574), and Switzerland (0.028118). Spain (-0.128713) and Norway (-0.336535) both decreased.**
